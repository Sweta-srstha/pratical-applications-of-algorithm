```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.minStack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        
        # change value for minimum for minStack
        val = min(val, self.minStack[-1] if self.minStack else val)
        self.minStack.append(val)

    def pop(self) -> None:
        self.stack.pop()
        self.minStack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.minStack[-1]


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

```python
class BrowserHistory:

    def __init__(self, homepage: str):
        self.history = [homepage]  # Initialize with the homepage
        self.current = 0  # Pointer to the current page

    def visit(self, url: str) -> None:
        # Move current pointer to the new URL and clear forward history
        self.current += 1
        self.history = self.history[:self.current]  # Clear forward history
        self.history.append(url)  # Add the new URL

    def back(self, steps: int) -> str:
        # Move back in history, but not beyond the start
        self.current = max(0, self.current - steps)  # Avoid going before index 0
        return self.history[self.current]  # Return the current URL

    def forward(self, steps: int) -> str:
        # Move forward in history, but not beyond the end
        self.current = min(len(self.history) - 1, self.current + steps)  # Avoid going beyond the last index
        return self.history[self.current]  # Return the current URL
```
